# Basics


*   **Prompt**: is the input or instruction a user provides to a generative AI model to get a specific output
*   **Prompt Engineering**: strategic process of crafting precise and effective instructions to guide generative AI models toward producing high-quality and relevant outputs

*   **Prompt Engineering Workflow**: iterative and test-driven process that can enhance model performance
image.png





## Example: A marketing campaign workflow
**Objective**: Get a new product description for an eco-friendly water bottle, targeting sustainability-conscious millennials.


---


**Initial Prompt:**
`Write a product description for an eco-friendly water bottle.`

**Evaluation**: It mentions basic features but doesn't appeal to the target audience.


---


**Refinement 1**: Add more context and a persona <br>
`You are a marketing expert for a sustainability-focused startup. Write a product description for an eco-friendly water bottle. The target audience is millennials who care about the environment.`

**Evaluation**: The output is better, using more appropriate language like "sustainability" and "conscious consumer." However, it's a long paragraph and lacks punch

---


**Refinement 2**: Specify the format, length, and tone <br>
`You are a marketing expert for a sustainability-focused startup. The target audience is millennials who care about the environment. Write three compelling product descriptions, each under 75 words, for an eco-friendly water bottle. Use an encouraging and inspirational tone.`

**Evaluation**: The outputs are now much closer to the goal. They are concise, targeted, and engaging.

---


**Refinement 3**: Improve clarity and provide a specific call to action <br>
`You are a marketing expert for a sustainability-focused startup. The target audience is millennials who care about the environment and want to reduce plastic waste. Write three compelling product descriptions, each under 75 words, for an eco-friendly water bottle. Use an encouraging and inspirational tone. For each one, include a short, powerful call to action like "Shop Now".`

**Evaluation**: The final outputs are highly effective, incorporating specific features and clear calls to action, directly addressing the initial objective.



# Components of a Prompt



1. ⭐**`Objective`**: What you want the model to achieve. Be specific and include any overarching objectives. Also called "mission" or "goal.


> Your objective is to help students with math problems without directly giving them the answer.

<br>

2.  ⭐**`Instructions`**: Step-by-step instructions on how to perform the task at hand. Also called "task," "steps," or "directions." *These are the directions you give inside your prompt.*


> 1. Understand what the problem is asking. <br>
> 2. Understand where the student is stuck. <br>
> 3. Give a hint for the next step of the problem. <br>

<br>

❗**`System Instructions`**
- These are rules/instructions set outside the prompt that may control or alter the model's behaviour.

- They apply across many tasks or prompts, not just one. For many model APIs, system instructions are specified in a dedicated parameter.
- Example: telling the model to “always reply politely and keep answers short”.
<br>

3.  **`Persona`**: Who or what the model is acting as. Also called "role" or "vision."


> You are a math tutor here to help students with their math homework

<br>

4. **`Constraints`**: Restrictions on what the model must adhere to when generating a response, including what the model can and can't do. Also called "guardrails," "boundaries," or "controls."


> Don't give the answer to the student directly. Instead, give hints at the next step towards solving the problem.

<br>

5. **`Tone`**: The tone of the response.


> Respond in a casual and technical manner.

<br>

6. **`Context`**: Any information that the model needs to refer to in order to perform the task at hand. Also called "background," "documents," or "input data."

> A copy of the student's lesson plans for math.

<br>

7. **`Few-shot examples`**: Examples of what the response should look like for a given prompt.


<br>


8. **`Reasoning steps`**: Tell the model to explain its reasoning. This can sometimes improve the model's reasoning capability. Also called "thinking steps."
> Explain your reasoning step-by-step.


<br>

9. **`Response format`**: The format that you want the response to be in. For example, you can tell the model to output the response in JSON.
> Format your response in JSON.

<br>

10. **`Recap`**: Concise repeat of the key points of the prompt, especially the constraints and response format, at the end of the prompt.
> Don't give away the answer and provide hints instead. Always format your response in JSON format.


<br>

11. **`Safeguards`**: Keeps the bot’s responses focused on its task. Also called safety rules.
>The bot must only provide guidance and hints related to solving math problems. It must not respond to questions outside math learning.

<br>

*Note: Depending on the specific tasks at hand, you might choose to include or exclude some of the optional components. You can also adjust the ordering of the components and check how that can affect the response.*

# Sample Prompt Template



```
<OBJECTIVE_AND_PERSONA>
You are a [insert a persona, such as a "math teacher" or "automotive expert"]. Your task is to...
</OBJECTIVE_AND_PERSONA>

<INSTRUCTIONS>
To complete the task, you need to follow these steps:
1.
2.
...
</INSTRUCTIONS>

------------- Optional Components ------------

<CONSTRAINTS>
Dos and don'ts for the following aspects
1. Dos
2. Don'ts
</CONSTRAINTS>

<CONTEXT>
The provided context
</CONTEXT>

<OUTPUT_FORMAT>
The output format must be
1.
2.
...
</OUTPUT_FORMAT>

<FEW_SHOT_EXAMPLES>
Here we provide some examples:
1. Example #1
Input:
Thoughts:
Output:
...
</FEW_SHOT_EXAMPLES>

<RECAP>
Re-emphasize the key aspects of the prompt, especially the constraints, output format, etc.
</RECAP>
```
